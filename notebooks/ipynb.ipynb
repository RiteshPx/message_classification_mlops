```python
import re
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
clean_text("This is a sample TEXT! Visit http://example.com.")
```
